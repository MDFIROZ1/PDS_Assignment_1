```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
data_frality = pd.read_excel("/content/FralityAnalysis.csv.xlsx")
data_frality
```

```python
data_frality.info()
```

```python
import seaborn as sns
```

```python
sns.lineplot(x='Age', y='Grip Strength', data=data_frality)
plt.title('Age Vs Grip Strength')
plt.xlabel('Age')
plt.ylabel('Grip Strength')
plt.savefig("/content/FralityAnalysis1.png")
plt.show()
```

```python
sns.scatterplot(x='Height', y='Grip Strength', data=data_frality)
plt.title('Height Vs Grip Strength')
plt.xlabel('Height')
plt.ylabel('Grip Strength')

# Save the plot as a PNG file
plt.savefig("/content/FralityAnalysis2.png")

# Display the plot
plt.show()
```

```python
sns.distplot(data_frality['Age'], kde = False, color ='green', bins = 50)
fig = plt.gcf()
fig.savefig("/content/FralityAnalysis3.png")
```

```python
data_scatter_plt = data_frality.plot.scatter(x='Weight',y='Grip Strength')
fig=data_scatter_plt.get_figure()
fig.savefig("/content/FralityAnalysis4.png")
```
